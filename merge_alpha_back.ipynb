```python
cont = 0

if not os.path.exists(back_ext):
    os.mkdir(back_ext)

if not os.path.exists(merged_folder):
    os.mkdir(merged_folder)



for folder in os.listdir(folder_with_images):
    if os.path.exists(f'{folder_with_images}//{folder}//interpolated_frames_film_a'):
        print(f'Resizing and merging folder: {folder}')
        for filename in os.listdir(f'{folder_with_images}//{folder}//interpolated_frames_film_a'):
            file = f'{folder_with_images}//{folder}//interpolated_frames_film_a//{filename}'
            img = cv2.imread(file, cv2.IMREAD_UNCHANGED)
            img = cv2.resize(img,(1920,1080))
            cv2.imwrite(f'{back_ext}//output_{cont:07d}.png',img)

            if os.path.exists(f'{alpha_folder}//output_{cont:07d}.png'):
                merge_bg(f'{lip_folder}//output_{cont:07d}.png', f'{alpha_folder}//output_{cont:07d}.png',
                         f'{back_ext}//output_{cont:07d}.png', f'{merged_folder}//output_{cont:07d}.png', 
                         offset_x=0, offset_y=760, w_res = 400, h_res = 320)

            cont = cont + 1
print('generating final video')
add_music_background(f'{full_folder}//full.wav',f'{MUSIC_BG_FOLDER}\\{MUSIC_BACKGROUND_NAME}',f'{full_folder}//full_bg.wav')
create_video_from_pngs(merged_folder, f'{full_folder}//full.mp4', music_path=f'{full_folder}//full_bg.wav')
```
